표준편차 : 분산의 루트

```python
import numpy as np
```

```python
data_01 = np.array([1,3,5,7,9])
data_02 = np.array([3,4,5,6,7])
```

```python
#data_01
((1-5)**2+(3-5)**2+(5-5)**2 + (7-5)**2 + (9-5)**2)/5
```

```python
#data_02 표준편차
np.sqrt(((1-5)**2+(3-5)**2+(5-5)**2 + (7-5)**2 + (9-5)**2)/5)
```
