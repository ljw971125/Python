```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud #  
from collections import Counter
from konlpy.tag import Okt
import numpy as np
#   
f=open('korealaw.txt','r',encoding='utf-8')
text=f.read()
#text

# 
okt=Okt() #  
nouns=okt.nouns(text) #    
#nouns

words=[n for n in nouns if len(n) >1 ] #   1  
c=Counter(words) #    

#  
wc=WordCloud(font_path='C:\\Windows\\Fonts\\NGULIM.TTF',background_color="white",width=800, height=400,relative_scaling=1).generate_from_frequencies(c)
plt.figure()
plt.imshow(wc)
```
