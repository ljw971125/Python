```python
pd.merge( df1, df2_right,on='이름', how='left') # 왼쪽 기준으로 병합
```

```python
pd.merge(df1,df2_right, on='이름', how='right') #오른쪽 기준으로 병합
```

```python
pd.merge(df1, df2_right, on='이름', how='inner') # 두 DataFrame에 모두 키 값이 존재하는 경우
```

```python
pd.merge(df1, df2_right, on='이름', how='outer') 
# 하나의 DataFrame에 키 값이 존재하는 경우 모두 병합
# Outer 방식에서는 없는 값은 NaN으로 대입됩니다.
```

```python
df2_right.columns = ['성함','연봉','가족수']
```

```python
df2_right
```

```python
pd.merge(df1, df2_right, left_on='이름', right_on= '성함', how='outer')
```
