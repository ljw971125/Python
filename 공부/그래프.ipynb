#그래프의 종류
plt.plot()  선 그래프
plt.bar() : 막대그래프
plt.scatter() : 산포도 그래프
plt.pie() : 타원형 그래프
plt.hist() : 히스토그램 그래프
plt.boxplot() : 박스플롯(주가창확인)
    
#그래프 설정에 필요한 사항
축레이블, 범례, 축범위, 선종류, 마커(표시), 색상, 
축스케일, 그리드, 눈금, 타이틀

```python
#막대그래프시각화
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
#그래프에 필요한 사항
#데이터(수치), 바탕 , 그래프화면(막대, x축, y축)
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#데이터
xdata=['AAA','BBB','CCC']
ydata=[10,20,30]

#바탕(그림)
fig=plt.figure()
#바탕->하위차트
fig.add_subplot(1,1,1) # 1행, 1열, 1개 그래프

#화면을 보여주기
plt.show()
```

```python
#그래프에 필요한 사항
#데이터(수치), 바탕(figure) , 바탕-그래프화면(막대, x축, y축)
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#데이터
xdata=['AAA','BBB','CCC']
ydata=[10,20,30]

#바탕(그림)
fig=plt.figure()
#바탕->하위차트->축을 만든다.
ax=fig.add_subplot(1,1,1) # 1행, 1열, 1개 그래프
#축을 만든 후 데이터입력, x,y축 설정
#ax.bar(xdata,ydata)
#ax.bar(range(len(xdata)),ydata)
ax.bar(xdata,ydata,align='center',color="darkblue")

#전체차트에서 필요한 사항은 plt에서 설정
plt.xlabel("NAME")
plt.ylabel("SCORE")
plt.title("CHART TITLE")

#그래프 완성 후 그림을 사용(저장)
plt.savefig('result.png')

#화면을 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
xdata = list(range(30))
ydata = []
for x in xdata:
    y = 2 * x
    ydata.append(y)
plt.plot(xdata, ydata, label='y=2x')
plt.xlim(0, 35) # x축의 범위
plt.ylim(0, 70) # y축의 범위
plt.xticks(list(range(0, 35 ,2))) #x축에서 중간표시 부분 0~35범위를 가지며 2씩 증가한 표시
plt.yticks(list(range(0, 70, 5))) #y축에서 중간표시 부분
plt.grid(True) #그래프의 격자 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
# 타원형의 그래프이며 표시할 데이터는 %단위로 계산된 값을 표시
xdata = ['개', '고양이', '기타', '기르지않음']
ydata = [18, 3.4, 3.1, 75.5 ] #수치 %단위
plt.pie(ydata, explode=(0, 0, 0, 0),labels=xdata, autopct='%1.2f%%',shadow=True, startangle=90)
# ydata : 타원형 그래프에 표시될 데이터
#explode=(0, 0, 0, 0) : 4개의 데이터 중 그래프를 튀어나오도록 표식
#labels=xdata : 타원형그래프에 label표시
#autopct='%1.2f%%' % %사이에 소수점 단위를 표시하기 위한 방법
# 1.2f%라는 문자열이 표시됨. 1.2는 소수 둘째자리까지 표시, f는 소수라는 표시
#shadow=True : 그래프를 입체적으로 표시
#startangle=90 : 시작데이터의 각도를 표시

plt.show()
```

```python
#하위차트를 2차원배열로 접근하기
x = list(range(1,11))
y = list(range(10, 101, 10))
#변수 선언은 1개 아래에서 2차원배열 활용하여 선택
fig, axs = plt.subplots(nrows=2, ncols=2,figsize=(9, 5), sharex=True, sharey=True)
ax = axs[0][0]
ax.plot(x,y)
ax.set_title('선 그래프 1')

ax = axs[0][1]
ax.plot(x,y, color='red', linestyle='--',marker='o')
ax.set_title('선 그래프 2')

ax = axs[1][0]
ax.bar(x,y)
ax.set_title('막대 그래프')

ax = axs[1][1]
ax.scatter(x,y)
ax.set_title('산포 그래프')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.2, hspace=0.3)
plt.show()
```
