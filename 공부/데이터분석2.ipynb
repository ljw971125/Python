```python
#문제 . 
#*****
#*****
#*****
#.....
#행과 열을 입력받아 행열크기의 사각별모양을 완성하시오.
#(함수로 작성)
def starsqrt(행,열):
    for row in range(0,행):
        for col in range(0,열):
            print("*",end="")
        print()

starsqrt(5,4)

def starsqrt(행,열,모양):
    for row in range(0,행):
        for col in range(0,열):
            print("{0}".format(모양),end="")
        print()

starsqrt(5,4,"$")
```

```python
# 평균기온 -> 전체합/일수
import csv
f = open('month_temp.csv', 'r', encoding='utf-8') 
lines = csv.reader(f)
next(lines) # 헤더를 건너뛴다.
sum = 0
day=0;
for line in lines:
    sum += float(line[2]) #전체일수의 누적 온도
    day+=1
    avg = sum/day 
print('평균 기온 : %.1f' %avg)
f.close()
```

```python
#원주시의 약국 중 5년이내 개설된 약국의 수는 몇개인가요?
import csv
f = open('pharm_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
header = next(lines)

wonjucnt=0
recent=0
for line in lines:
    if line[1] == '원주시':
        wonjucnt=wonjucnt+1
        if int(line[3]) > 20140901 :
            recent += 1
    
#표현언어 : 원주시의 총 약국수는 %d개이며 5년이내 개설된 약국수는 %d개입니다.
print("원주시의 총 약국수는 %d개이며 5년이내 개설된 약국수는 %d개입니다."%(wonjucnt,recent))
```
