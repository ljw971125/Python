```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('20230313.csv')

data = csv.reader(f)
next(data) # header    

max_temp = [] #     

for row in data:
    if row[-1] !='':
        max_temp.append(float(row[-1]))


plt.figure(figsize=(10,2),dpi=300)
plt.plot(max_temp,'r')
plt.show()
f.close()
```

```python
#
# 04. (12 25)      .
#   - : hotpink, : high
#   - : skyblue, : low
#  :   
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] #    
min_temp = []

for row in data:
    if row[-1] != '' and row[-2] != '': #   
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25': # 12 25 
            max_temp.append(float(row[4])) #    
            min_temp.append(float(row[-2])) #    

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('  ')
plt.plot(max_temp,'hotpink', label = 'high')
plt.plot(min_temp,'skyblue', label = 'low')

plt.legend()
plt.show()

f.close()
```

```python
#
# 04. 8    1      .
#   - : red, : aug
#   - : blue, : jan
#  : 1 8  
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] #    
min_temp = [] #    

for row in data:
    if row[-1] != '' and row[-2] != '': #   
        if row[0].split('-')[1] == '08': # 12 25 
            max_temp.append(float(row[4])) #    
        elif row[0].split('-')[1] == '01':
            min_temp.append(float(row[-2])) #    

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('  ')
plt.hist(max_temp, bins=100, color='r', label='aug') 
plt.hist(min_temp, bins=100, color='b', label='jen')

plt.legend()
plt.show()

f.close()
```
