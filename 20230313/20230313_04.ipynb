```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('20230313.csv')

data = csv.reader(f)
next(data) # header를 제외한 데이터를 가져오기 위함

max_temp = [] # 최고 기온 데이터를 저장할 리스트

for row in data:
    if row[-1] !='':
        max_temp.append(float(row[-1]))


plt.figure(figsize=(10,2),dpi=300)
plt.plot(max_temp,'r')
plt.show()
f.close()
```

```python
# 03. 크리스마스(12월 25일)의 최고 기온을 그래프로 표현해주세요.
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] # 최고 기온을 저장할 리스트

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25': # 12월 이라면
            max_temp.append(float(row[4])) # 리스트에 최고 기온값을 추가

plt.plot(max_temp,'hotpink')

f.close()
```

```python
#본인성명
# 04. 크리스마스(12월 25일)의 최고 기온과 최저 기온을 그래프로 표현해주세요.
# 최고 기온 - 색상: hotpink, 범례: high
# 최저 기온 - 색상: skyblue, 범례: low
# 제목 : 크리스마스 기온 변화
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] # 최고 기온을 저장할 리스트
min_temp = []

for row in data:
    if row[-1] != '' and row[-2] != '': # 값이 비어있을 경우
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25': # 12월 25일 이라면
            max_temp.append(float(row[4])) # 리스트에 최고 기온값을 추가
            min_temp.append(float(row[-2])) # 리스트에 최저 기온값을 추가

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('크리스마스 기온 변화')
plt.plot(max_temp,'hotpink', label = 'high')
plt.plot(min_temp,'skyblue', label = 'low')

plt.legend()
plt.show()

f.close()
```

```python
# 8월의 최고기온 데이터를 히스토그램으로 표현하기

import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] # 최고 기온을 저장할 리스트

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            max_temp.append(float(row[4]))

plt.hist(max_temp, bins=100, color='r')
plt.show()
f.close()
```

```python
# 1월의 최저기온 데이터를 히스토그램으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

min_temp = [] # 최고 기온을 저장할 리스트

for row in data:
    month = row[0].split('-')[1]
    if row[-2] != '':
        if month == '01':
            min_temp.append(float(row[-2]))

plt.hist(min_temp, bins=100, color='r')
plt.show()
f.close()
```

```python
#이지운
# 04. 8월의 의 최고 기온과 1월의  최저 기온을 히스토그램 그래프로 표현해주세요.
# 최고 기온 - 색상: red, 범례: aug
# 최저 기온 - 색상: blue, 범례: jan
# 제목 : 1월과 8월의 기온 변화
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')

data = csv.reader(f)
next(data)

max_temp = [] # 최고 기온을 저장할 리스트
min_temp = [] # 최저 기온을 저장할 리스트

for row in data:
    if row[-1] != '' and row[-2] != '': # 값이 비어있을 경우
        if row[0].split('-')[1] == '08': # 12월 25일 이라면
            max_temp.append(float(row[4])) # 리스트에 최고 기온값을 추가
        elif row[0].split('-')[1] == '01':
            min_temp.append(float(row[-2])) # 리스트에 최저 기온값을 추가

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('크리스마스 기온 변화')
plt.hist(max_temp, bins=100, color='r', label='aug') 
plt.hist(min_temp, bins=100, color='b', label='jen')

plt.legend()
plt.show()

f.close()
```
