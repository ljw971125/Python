```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='Malgun Gothic')
```

```python
plt.rcParams["figure.figsize"] = (12,9)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
sns.violinplot(x='day', y='total_bill',data=tips)
plt.title('violin plot')
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill',data=tips)
plt.title('replot')
plt.show()
```

```python
sns.lmplot(x='tip', y='total_bill',data=tips)
plt.title('lmplot')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, linewidths=1)
plt.title('heatmap')
plt.show()
```

```python
plt.bar(tips['day'], tips['total_bill'])
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='Oranges')
```

```python
sns.palplot(sns.light_palette((210,90,60), input='husl'))
```

```python
sns.palplot(sns.dark_palette('muted purple', input='xkcd'))
```

```python
sns.palplot(sns.color_palette('BrBG',10))
```

```python
sns.palplot(sns.color_palette('BrBG_r',10))
```

```python
sns.palplot(sns.color_palette('coolwarm',10))
```

```python
sns.barplot(x='tip', y='total_bill', data = tips, palette='coolwarm')
```

```python
sns.barplot(x='tip', y='total_bill', data = tips, palette='Reds')
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker',
            data=tips,
            kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker',
            col='time',
            data=tips,
            kind='bar')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 1000
```

```python
plt.scatter(x, y, s=area, c = colors)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(131)
plt.scatter(x, y, s=area, c='blue', alpha=0.1)
plt.title('alpha 0.1')

plt.subplot(132)
plt.scatter(x, y, s=area, c='red', alpha=0.5)
plt.title('alpha 0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, c='green', alpha=1.0)
plt.title('alpha 1.0')

plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E', 'F']
y = [66, 80, 60, 50, 80, 10]

plt.bar(x, y, align='center', alpha = 0.7, color = 'red')

plt.xticks(x)
plt.ylabel('Scores')
plt.title('Subjects')

plt.show()
```
