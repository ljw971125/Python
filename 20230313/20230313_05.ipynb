```python
import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')
data = csv.reader(f)

next(data)
jan_min = [] # 1월의 최고기온 데이터
aug_max = [] # 8월의 최고기온 데이터

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' and row[-2] != '':
        if month == '08':
            aug_max.append(float(row[4]))
        if month == '01':
            jan_min.append(float(row[3]))
    
plt.boxplot([jan_min, aug_max])
plt.show()

f.close()
```

```python
# 이지운
# 06. 월별 최고기온 데이터를 박스플롯으로 표현해주세요.

import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')
data = csv.reader(f)

next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[4]))

plt.boxplot(month, showfliers=False)
plt.show()


f.close()
```

```python
# 8월의 일자별 최고기온 데이터를 박스플롯으로 표현해주세요.

import csv
import matplotlib.pyplot as plt

f = open('20230313.csv')
data = csv.reader(f)

next(data)
day = []
for i in range(31):
    day.append([])
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[3]))

plt.figure(figsize=(10,5),dpi = 300)
plt.boxplot(day, showfliers=False)
plt.show()

f.close()
```
