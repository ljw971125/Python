```python
f = open('20230314_01.csv')
data = csv.reader(f)

age_result = []

for row in data:
    if '신도림' in row[0]: # 1행이 신도림이면
        for i in row[3:]: # 
            age_result.append(int(i))

print(age_result)

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title("신도림동의 인구 구조")
plt.plot(age_result)
plt.show()

f.close()
```

```python
f = open('20230314_01.csv')
data = csv.reader(f)

age_result = []

for row in data:
    if '신도림' in row[0]: # 1행이 신도림이면
        for i in row[3:]: # 
            age_result.append(int(i))

print(age_result)

plt.figure(figsize=(10,5), dpi=300)
plt.title("신도림동의 인구 구조")
plt.bar(range(101),age_result)
plt.show()

f.close()
```
