```python
import csv
import matplotlib.pyplot as plt
```

```python
plt.pie([10,20])
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]

label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'skyblue', 'pink']
plt.rc('font', family='Malgun Gothic')

plt.axis('equal') # 원이 찌그러 졌을때
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0)) #소수점 표시
plt.legend()
plt.show()
```

```python
f = open('20230314_02.csv')
data = csv.reader(f)
size = []

for row in data:
    if '신도림' in row[0]:
        male = 0
        female = 0
        for i in range(101):
            male += int(row[i+3])
            female += int(row[i+106])
        break

size.append(male)
size.append(female)

color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남','여'],autopct='%.1f%%',colors=color, startangle=90)
plt.title('신도림동의 남녀 성별 비율')
plt.legend()
plt.show()
```
