```python
f = open('20230314_02.csv')
data = csv.reader(f)

male = []
female = []

for row in data:
    if '신도림' in row[0]:
        for i in range(3, 104): #
            male.append(int(row[i]))
        for i in range(106, 207): # 
            female.append(int(row[i]))

print(female)

f.close()

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title("신도림동의 남녀 성별 인구 분포")
plt.barh(range(101),male,label='male')
plt.barh(range(101),female,label='female')
plt.legend()
plt.show()

f.close()
```

```python
f = open('20230314_02.csv')
data = csv.reader(f)

male = []
female = []

for row in data:
    if '신도림' in row[0]:
        for i in range(3, 104): 
            male.append(-int(row[i]))
        for i in range(106, 207): 
            female.append(int(row[i]))

print(female)

f.close()

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,5), dpi=300)
plt.title("신도림동의 남녀 성별 인구 분포")
plt.barh(range(101),male,label='male')
plt.barh(range(101),female,label='female')
plt.legend()
plt.show()

f.close()
```
